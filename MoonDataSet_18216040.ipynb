# Import Modules

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_validate
```

# Read dataset
Membaca csv yang telah disediakan

```python
train = pd.read_csv('el4233-2018-2019-02-klasifikasi-train.csv', index_col=0)
test = pd.read_csv('el4233-2018-2019-02-klasifikasi-test.csv', index_col=0)
```

# Split dataset

```python
x = train.drop('Y',axis=1) #memisahkan Y, hanya menanmpilkan X0 dan X1
y = train['Y']
```

# Plot dataset
Melihat persebaran data dan menampilkan gambarnya

```python
plt.close('all')

plt.figure(figsize=(10, 6))

zero = train['Y'] == 0
one = train['Y'] == 1
train_zero = train[zero]
train_one = train[one]

plt.plot(train_zero['X0'],train_zero['X1'],"r.")
plt.plot(train_one['X0'],train_one['X1'],"c*")

plt.xlabel("X0", fontsize=14)
plt.ylabel("X1", fontsize=14)
plt.axis([-1.6, 2.6, -1.6, 2.6])
plt.show()
```
